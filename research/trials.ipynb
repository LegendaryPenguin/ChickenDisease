```python
def get_prodict(x:int, y:int) -> int:
    return x*y
```

```python
get_prodict(x=2, y=4)
```

```python
get_prodict(x=2, y="4")
```

```python
import unittest

if not hasattr(unittest.TestCase, "assertRaisesRegexp"):
    unittest.TestCase.assertRaisesRegexp = unittest.TestCase.assertRaisesRegex

from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_prodict(x:int, y:int) -> int:
    return x*y
```

```python
get_prodict(x=2, y=4)
```

```python
try:
    get_prodict(x=2, y="4")
except Exception as e:
    print("ensure caught bad type:", type(e).__name__)
```
